### CVE Deep Dive: SSL POODLE (CVE-2014-3566)
- **Description:** A padding oracle vulnerability in SSLv3 that allows attackers to decrypt secure HTTPS traffic.
- **Impact:** Enables man-in-the-middle decryption of SSL traffic under certain conditions.
- **Mitigation:** Disable SSLv3 and enable only TLS 1.2 or above.
- **Verification:** Run Nmap with the `--script ssl-enum-ciphers` flag to ensure SSLv3 is disabled.

### SSL/TLS Cipher Analysis (Port 443)
The Nmap `ssl-enum-ciphers` scan on host 192.168.1.1 identified supported encryption protocols and ciphers.

**Findings:**
- Older SSLv3 protocol is enabled, confirming exposure to the **POODLE vulnerability (CVE-2014-3566)**.
- Several weak cipher suites (e.g., `TLS_RSA_WITH_AES_128_CBC_SHA`) are still accepted.
- The server does not enforce modern encryption standards (TLS 1.2/1.3).

**Impact:**
These weaknesses allow potential downgrade and man-in-the-middle attacks, where encrypted traffic could be decrypted.

**Mitigation Recommendations:**
- Disable SSLv3 completely in server configuration.
- Restrict accepted cipher suites to strong modern algorithms (AES-GCM, SHA256, TLS 1.2+).
- Periodically test SSL configurations with online tools like **Qualys SSL Labs**.

```python
# Day8 Step (Fixed): Automated Risk Correlation from Parsed Services
import pandas as pd
from pathlib import Path

# Load parsed CSV
csv_path = Path("../raw_outputs/Day8/parsed_services_fixed_192_168_1_1.csv")
if not csv_path.exists():
    print("⚠️ Parsed file not found!")
else:
    df = pd.read_csv(csv_path)
    risks = []

    for _, row in df.iterrows():
        svc = str(row['service']).lower()
        info = str(row['info']).lower()
        risk = "Unknown"
        severity = "Low"
        mitigation = "N/A"

        # --- Rule-based mapping ---
        if "ftp" in svc:
            risk, severity, mitigation = (
                "Weak authentication risk",
                "Medium",
                "Use FTPS or disable if unused"
            )
        elif "http" in svc and "ssl" not in svc:
            risk, severity, mitigation = (
                "Outdated web server may allow XSS/RCE",
                "High",
                "Upgrade to supported version"
            )
        elif "https" in svc or "ssl" in info:
            risk, severity, mitigation = (
                "Weak SSL/TLS version",
                "High",
                "Enforce TLS 1.2 or higher"
            )
        elif "dns" in svc:
            risk, severity, mitigation = (
                "DNS cache poisoning risk",
                "Medium",
                "Restrict recursion and update daemon"
            )
        elif "freeciv" in svc:
            risk, severity, mitigation = (
                "Unnecessary open service",
                "Medium",
                "Close or restrict this port"
            )

        risks.append({
            "Port": row['port'],
            "Service": row['service'],
            "Version/Info": row['info'],
            "Risk": risk,
            "Severity": severity,
            "Recommended Mitigation": mitigation
        })

    risk_df = pd.DataFrame(risks)
    output_path = Path("../raw_outputs/Day8/service_risk_map.csv")
    risk_df.to_csv(output_path, index=False)

    print(f"✅ Risk correlation completed. Saved to: {output_path}")
    display(risk_df)
```

```python
import os
from pathlib import Path
import pandas as pd

# Define paths
base_path = Path("../raw_outputs")
hosts = ["192_168_1_1", "192_168_1_2", "192_168_1_4", "192_168_1_7"]

# Function to parse Nmap files
def parse_nmap_text(filepath):
    data = []
    if not filepath.exists():
        return data
    with open(filepath, "r") as f:
        for line in f:
            if "/tcp" in line and "open" in line:
                parts = line.strip().split()
                port = parts[0]
                service = parts[2] if len(parts) > 2 else "unknown"
                data.append({"File": filepath.name, "Port": port, "Service": service})
    return data

# Load all host scans
all_data = []
for host in hosts:
    for day in ["Day2", "Day3", "Day4"]:
        file = base_path / day / f"detailed_scan_{host}.txt"
        if not file.exists():
            file = base_path / day / f"vuln_scan_{host}.txt"
        all_data.extend(parse_nmap_text(file))

df_compare = pd.DataFrame(all_data)
df_compare.to_csv("../raw_outputs/Day9/host_port_summary.csv", index=False)
df_compare.head(10)
```

```python
# Day9 Step 3: Create Cross-Host Service Comparison Table

# Pivot the data for clarity
pivot_df = df_compare.pivot_table(
    index="Port",
    columns="File",
    values="Service",
    aggfunc=lambda x: ", ".join(set(x))
).fillna("—")

# Save the comparison result
pivot_df.to_csv("../raw_outputs/Day9/cross_host_comparison.csv")

pivot_df.head(15)
```

```python
# ✅ Day9 Step 4 (Fixed): Identify Common and Unique Ports Across Hosts

# Use df_compare consistently
hosts_ports = {
    file: set(df_compare[df_compare["File"] == file]["Port"])
    for file in df_compare["File"].unique()
}

# Find common ports across all hosts
common_ports = set.intersection(*hosts_ports.values())

# Find unique ports per host
unique_ports = {file: ports - common_ports for file, ports in hosts_ports.items()}

print("=== Common Ports Across All Hosts ===")
print(common_ports if common_ports else "No common ports found")

print("\n=== Unique Ports Per Host ===")
for file, ports in unique_ports.items():
    print(f"\n{file}: {ports if ports else 'No unique ports'}")
```

```python
# Day10 Step 2: Extract banner and version details from all Nmap scan files

from pathlib import Path
import pandas as pd
import re

# Define path for Day10 and previous scans
base_dir = Path("../raw_outputs")
scan_files = list(base_dir.glob("Day3/*.txt")) + list(base_dir.glob("Day4/*.txt"))

data = []

# Regex to extract Port / Service / Version
pattern = re.compile(r"(\d{1,5}/tcp)\s+open\s+([\w-]+)(?:\s+([\w\d\.\-_/]+))?")

for file in scan_files:
    with open(file, "r") as f:
        content = f.read()
        matches = pattern.findall(content)
        for port, service, version in matches:
            data.append({
                "File": file.name,
                "Port": port,
                "Service": service,
                "Version": version if version else "N/A"
            })

# Create DataFrame
df_banners = pd.DataFrame(data)
df_banners.to_csv("../raw_outputs/Day10/banner_extract.csv", index=False)

df_banners.head(10)
```

```python
# Day10 Step 3: Flag Outdated or Risky Service Versions Automatically

# Define known risky/outdated software versions (simple baseline check)
risky_versions = {
    "Bftpd": "2.2",
    "dnsmasq": "2.78",
    "Boa": "0.94.13",
    "OpenSSL": "1.0.1i",
    "Freeciv": "Unknown"
}

def check_risk(row):
    for key, ver in risky_versions.items():
        if key.lower() in row["Service"].lower() or key.lower() in row["Version"].lower():
            return f"⚠️ Possible outdated version ({key} {ver})"
    return "✅ No known issue detected"

# Apply the risk check
df_banners["Risk_Assessment"] = df_banners.apply(check_risk, axis=1)

# Save the risk-assessed data
df_banners.to_csv("../raw_outputs/Day10/banner_risk_analysis.csv", index=False)

# Show top entries
df_banners.head(15)
```

```python
# Day13: Parse ssl-enum-ciphers result into a CSV (one-line-per-finding)
from pathlib import Path
import re
import pandas as pd

infile = Path("../raw_outputs/Day13/ssl_enum_192_168_1_1.txt")
outfile = Path("../raw_outputs/Day13/ssl_enum_192_168_1_1_parsed.csv")

if not infile.exists():
    print("ERROR: ssl enum file not found:", infile)
else:
    lines = infile.read_text(errors="ignore").splitlines()
    findings = []
    cur_proto = None
    for L in lines:
        Ls = L.strip()
        # record protocol lines e.g. "TLSv1.2:"
        m = re.match(r'^(TLSv\d(\.\d)?)[:\s]*$', Ls)
        if m:
            cur_proto = m.group(1)
            continue
        # capture cipher lines like "TLS_RSA_WITH_AES_128_CBC_SHA - " or "Preferred: yes"
        if "Cipher" in Ls or re.search(r'^[A-Z0-9_]+_[A-Z0-9_]+', Ls):
            findings.append({"Protocol": cur_proto or "", "Line": Ls})
        # catch VULNERABLE / POODLE / SSLv3 notes
        if re.search(r'vulnerable|POODLE|SSLv3|sslv2|CVE-', Ls, re.I):
            findings.append({"Protocol": cur_proto or "", "Line": Ls})

    if not findings:
        print("No structured SSL findings parsed (file may have different format). Showing raw head:")
        print("\n".join(lines[:60]))
    else:
        df = pd.DataFrame(findings)
        df.to_csv(outfile, index=False)
        print(f"✅ Parsed {len(df)} ssl findings → {outfile}")
        display(df.head(20))
```

```python
# Day14 Step 3: Clean and normalize the consolidated dataset
import pandas as pd
from pathlib import Path

input_file = Path("../raw_outputs/Day14/master_consolidated_all.csv")
output_file = Path("../raw_outputs/Day14/master_cleaned.csv")

if not input_file.exists():
    print("⚠️ Master file not found. Run consolidation first.")
else:
    df = pd.read_csv(input_file)

    # Standardize column names
    df.columns = [c.strip().replace(" ", "_").replace("-", "_").lower() for c in df.columns]

    # Drop completely empty columns
    df = df.dropna(how="all", axis=1)

    # Fill missing values with "N/A"
    df = df.fillna("N/A")

    # Remove duplicate rows
    before = len(df)
    df = df.drop_duplicates().reset_index(drop=True)
    after = len(df)

    # Save cleaned version
    df.to_csv(output_file, index=False)

    print(f"✅ Cleaned dataset saved: {output_file}")
    print(f"📊 {before - after} duplicates removed | {len(df.columns)} columns | {len(df)} rows remaining")
    print("\n🧩 Columns after cleaning:")
    print(list(df.columns))
    display(df.head(10))
```

```python
# Day14 Step 4: Summarize risk levels and visualize
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

data = Path("../raw_outputs/Day14/master_cleaned.csv")

if not data.exists():
    print("⚠️ Cleaned file not found. Run previous step first.")
else:
    df = pd.read_csv(data)

    # If column names differ, we adapt dynamically
    cols = df.columns
    text_cols = [c for c in cols if df[c].dtype == "object"]

    # Combine all text columns into one large searchable text
    df["combined"] = df[text_cols].astype(str).agg(" ".join, axis=1)

    # Simple keyword-based tagging
    def classify(text):
        t = text.lower()
        if any(k in t for k in ["vulnerable", "cve", "exploit", "ssl", "poodle", "weak cipher"]):
            return "High Risk"
        elif any(k in t for k in ["ftp", "http", "open", "version", "service"]):
            return "Medium Risk"
        elif any(k in t for k in ["info", "scan", "port", "discovered"]):
            return "Low Risk"
        return "Unclassified"

    df["risk_level"] = df["combined"].apply(classify)
    summary = df["risk_level"].value_counts()

    # Visualization
    plt.figure(figsize=(6, 4))
    summary.plot(kind="bar", color=["red", "orange", "yellow", "gray"], edgecolor="black")
    plt.title("Risk Level Distribution - Consolidated Scan Data")
    plt.xlabel("Risk Category")
    plt.ylabel("Count of Findings")
    plt.grid(axis="y", linestyle="--", alpha=0.6)
    plt.tight_layout()
    plt.show()

    print("\n=== Risk Summary ===")
    print(summary)
```
